**Shift methtod to calculate relative change wrf to another**

```python
AAPL = close[["AAPL"]]
AAPL.head()
```

```python
AAPL["lag1"] = AAPL.shift(periods = 1)
AAPL.head()
```

```python
AAPL["dif-lag1"] = AAPL.AAPL - AAPL.lag1
AAPL
```

```python
AAPL["pct_change_lag1"] = AAPL.AAPL.div(AAPL.lag1).sub(1).mul(100)
AAPL
```

**Absolute change and relative change**

**diff() and pct_change()**


```python
AAPL2 = close[["AAPL"]]
AAPL2["dif-lag1"] = AAPL2.AAPL.diff(periods = 1)
AAPL2["pct_change_lag1"] = AAPL2.AAPL.pct_change(periods = 1)
AAPL2.head()
```

```python
AAPL2.AAPL.resample("M").last().pct_change(periods = 1).mul(100)
```

```python
ret = close.pct_change().dropna()
```

```python
ret.head()
```

```python
ret.describe()
```

```python
ret.describe().T
```

```python
summary = ret.describe().T[["mean", "std"]]
print(summary)
summary["mean"] = summary["mean"] * 252
summary["std"] = summary["std"] * np.sqrt(252)
print(summary)
```

```python
summary.plot.scatter("std", "mean", figsize=(12,8), fontsize=15)
for i in summary.index:
  plt.annotate(i, xy=(summary.loc[i, "std"]+0.002, summary.loc[i, "mean"]+0.002), size = 15)
plt.xlabel("Ann. Risk (std)", fontsize = 15)
plt.ylabel("Ann. Return", fontsize = 15)
plt.title("Risk/Return", fontsize = 15)
plt.show()
```

**Simple Moving Averages with rolling() function**



```python
plt.style.use("seaborn")
```

```python
SP500 = pd.read_csv("https://raw.githubusercontent.com/nadh1981/colabnotes/main/udemy-practice/python-algo-trading-a-z/financial-data-analysis/Video_Lecture_NBs/SP500.csv?token=AAHIWAE6ZQOSTWYBZHXEPPTA6C4FQ", parse_dates=["Date"], index_col="Date")
```

```python
SP500Close = SP500[["Close"]]
SP500Close = SP500Close.loc["2010-01-01":"2022-06-30"]
SP500Close.head()
```

```python
SP500Close.plot(figsize=(12,8), fontsize=15)
plt.legend(loc = "upper left", fontsize = 15)
plt.show()
```

```python
SP500Close["SMA10"] = SP500Close['Close'].rolling(window = 10).mean()
SP500Close["SMA20"] = SP500Close['Close'].rolling(window = 20).mean()
SP500Close["SMA50"] = SP500Close['Close'].rolling(window = 50).mean()
SP500Close["SMA200"] = SP500Close['Close'].rolling(window = 200).mean()
SP500Close.tail()
```

```python
SP500Close[["SMA50", "SMA200"]].plot(figsize=(12,8), fontsize=15)
plt.legend(loc = "upper left", fontsize = 15)
plt.show()
```



---
**Exponential weighted moving averates (EWMA)**


```python
SP500Close["EWM200"] = SP500Close[["Close"]].ewm(span = 200, min_periods=200).mean()
SP500Close["EWM50"] = SP500Close[["Close"]].ewm(span = 50, min_periods=50).mean()
```

```python
SP500Close[["SMA50", "EWM50"]].plot(figsize=(12,8`), fontsize=15)
plt.legend(loc = "upper left", fontsize = 15)
plt.show()
```
